```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
test_noisy = np.load("./data/test_noisy_100.npy")
train_noisy = np.load("./data/train_noisy.npy")
train_clean = np.load("./data/train_clean.npy")
```

## Data Shape

```python
print("test_noisy", test_noisy.shape)
print("train_noisy", train_noisy.shape)
print("train_clean", train_clean.shape)
```

## test_noisy

```python
n = 5

fig, ax = plt.subplots(1,n, figsize=(n*3,5))
ints = np.random.randint(0, high=test_noisy.shape[0], size=n)
for i in range(n):
    ax[i].imshow(test_noisy[ints[i]])

plt.show()
```

## train_noisy
